```python
import numpy as np
import pandas as pd
```

```python
iter_file = 'binary-untargeted-randomAscent-0.5-3-iter-19/loss_history.npy'
once_file = 'binary-untargeted-randomAscent-0.5-3-once-18/loss_history.npy'
none_file = 'binary-untargeted-randomAscent-0.5-3-none-18/loss_history.npy'
lh_iter = np.load(iter_file)
lh_once = np.load(once_file)
lh_none = np.load(none_file)
```

```python
import matplotlib.pyplot as plt
plt.plot(lh_none, label='Full')
plt.plot(lh_once, label='Randomly Sample Once')
plt.plot(lh_iter, label='Randomly Sample Every Iter')
plt.ylabel('Average Hinge Loss')
plt.xlabel('Number of Iterations')
plt.title('Hinge Loss per Iter')
plt.legend()
plt.show()
```
